```python
import cv2
import numpy as np,pandas as pd,os
```

```python
infu=r"D:\projects and learning\ai_ocr\val\images"#input folder name
onfu=r"D:\projects and learning\ai_ocr\val\labels"#output folder name
annoti="0"
```

```python
items=os.listdir(infu)
x_coori=0
y_coori=0
lbd=0
lbu=0
m_move=0
x_temp1=0
y_temp1=0
start=0
```

```python
def mouse_callback(event, x, y,flags,param):
    global x_coori,y_coori,lbd,lbu
    if (event == cv2.EVENT_LBUTTONDOWN):
        lbd=1
        lbu=0
        x_coori=x
        y_coori=y
    elif (event == cv2.EVENT_LBUTTONUP):
        lbu=1
        lbd=0
        x_coori=x
        y_coori=y
    elif (event==cv2.EVENT_MOUSEMOVE):
        x_coori=x
        y_coori=y
def rotate(image,location):
    print(image.shape)
    image=cv2.rotate(image,rotateCode = cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(location,image)
    return image
def all_rectangle(image):
    img=cv2.imread("option.png")
    img=cv2.resize(img,(100,800))
    image=np.hstack((img,image))
    return image
```

```python
for files in items:
    file_path=infu+"\\"+files
    name, _ = os.path.splitext(files)
    name=onfu+"\\"+name+".txt"
    image=cv2.imread(file_path)
    image=cv2.resize(image,(800,800))
    h,w,_=image.shape
    save_img=True
    initial="yes"
    capture=0
    img_true=image.copy()
    cv2.imwrite("delete.png",image)
    need=1
    while(True):
        image=cv2.imread("delete.png")
        if (need==1):
            image=all_rectangle(image)
            cv2.imwrite("delete.png",image)
            need=0
            continue
        cv2.namedWindow("image")
        cv2.setMouseCallback("image", mouse_callback)
        if (lbd==1):
            if(0<x_coori<100):
                lbd=0
                print("hii")
                if(0<y_coori<133):#start
                    start=1
                elif(133<y_coori<266):#next
                    start=1
                    break
                elif (266<y_coori<400):#retake
                    with open(name, "w") as file:
                        file.close()
                    start=1
                    cv2.imwrite("delete.png",img_true)
                    file=open("_.txt", "w")
                    need=1
                    continue
                elif(400<y_coori<533):#skip
                    save_img=False
                    blank=np.ones((800,800,3)).astype(np.uint8)
                    cv2.imwrite(file_path,blank)
                    break
                elif(533<y_coori<666):
                    image=rotate(img_true,file_path)
                    img_true=image
                    cv2.imwrite("delete.png",image)
                    need=1
                    continue
        if (start==1):
            if (lbd==1 and initial=="yes"):
                x_temp1,y_temp1=x_coori,y_coori
                 
                initial="no"
             
            if (lbu==1 and initial=="no"):
                lbu=0
                initial="yes"
                x_temp2,y_temp2=x_coori,y_coori
                capture=1
        if (initial=="no"):
            cv2.rectangle(image,(x_temp1,y_temp1),(x_coori,y_coori),(0,255,0),3)
            
        if (capture==1):
            cv2.rectangle(image,(x_temp1,y_temp1),(x_coori,y_coori),(0,255,0),3)
            cv2.imwrite("delete.png",image)
            x_mean_nor=(x_temp1+x_temp2-200)/(2*w)
            y_mean_nor=(y_temp1+y_temp2)/(2*h)
            width_nor=np.abs(x_temp1-x_temp2)/w
            height_nor=np.abs(y_temp1-y_temp2)/h
            to_write=f"{annoti} {x_mean_nor} {y_mean_nor} {width_nor} {height_nor}\n"
            print(to_write)
            with open(name, "a") as file:
                file.write(to_write)
            capture=0
        cv2.imshow("image",image)    
        cv2.waitKey(1)
    try:    
        file.close()
    except:
        pass
    cv2.waitKey(1)
cv2.destroyAllWindows()   
```
